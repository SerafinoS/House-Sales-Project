```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(df['sqft_living'].to_frame(),df['price'])
model.score(df['sqft_living'].to_frame(),df['price'])
```

```python
dataframe = df[['floors','waterfront','lat','bedrooms','sqft_basement','view','bathrooms','sqft_living15','sqft_above','grade','sqft_living']]
model = LinearRegression().fit(dataframe,df['price'])
model.score(dataframe,df['price'])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2)
model = LinearRegression()
pipe = Pipeline([('polynomial', poly), ('regressor', model)])
pipe.fit(dataframe,df['price']).score(dataframe,df['price'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge

dataframe = df[['price','floors','waterfront','lat','bedrooms','sqft_basement','view','bathrooms','sqft_living15','sqft_above','grade','sqft_living']]
ridge = Ridge(alpha=0.1)
X_train,X_test,y_train,y_test = train_test_split(dataframe.drop('price',axis=1),dataframe['price'])

ridge.fit(X_train,y_train)
ridge.score(X_test,y_test)
```

```python
poly = PolynomialFeatures(2)
test_poly= poly.fit_transform(X_test)
train_poly= poly.fit_transform(X_train)

ridge = Ridge(alpha=0.1)
ridge.fit(train_poly,y_train)
ridge.score(test_poly,y_test)
```
